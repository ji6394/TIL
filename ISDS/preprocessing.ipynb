# ITS

```python
import statsmodels.api as sm
import numpy as np
#ITS(Intervention Time Series Analysis)
# 2023년을 기점으로 개입 변수 생성
k1['쿠팡'] = k1['연도'].apply(lambda x: 1 if int(x) > 2023 else 0)
k2['쿠팡'] = k2['연도'].apply(lambda x: 1 if int(x) > 2023 else 0)
```

```python
k1
```

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
# '관중수'에 대한 회귀분석 준비
X = sm.add_constant(k1['쿠팡'])  # 상수항 추가 및 '쿠팡'이 개입 변수 역할을 합니다.
y = k1['관중수']

# OLS 회귀모델 구축 및 결과 출력
model = sm.OLS(y, X).fit()
print(model.summary())
```

R-squared (결정계수): 0.020입니다. 이는 '쿠팡' 변수가 관중 수 변동성의 약 2%만을 설명한다는 것을 의미합니다. 대체로, 결정계수가 낮다는 것은 모델이 데이터에 잘 맞지 않는다는 것을 나타냅니다.

F-statistic (F-통계량) 및 Prob (F-statistic): F-통계량은 39.57이며, 그에 대한 p-value는 3.92e-10입니다. 이는 모델이 통계적으로 유의하다는 것을 의미합니다. 즉, '쿠팡' 변수는 관중 수에 대해 통계적으로 유의미한 영향을 미칩니다.

계수 (coef): '쿠팡' 변수의 계수는 9105.4005입니다. 이는 '쿠팡' 변수가 1 단위 증가할 때마다 관중 수가 평균적으로 9105명 증가한다는 것을 의미합니다. 상수항(const)의 계수는 6161.8218로, 이는 '쿠팡' 변수의 영향을 받지 않는 기본 관중 수를 나타냅니다.

P>|t|: '쿠팡' 변수에 대한 p-value는 0.000으로, 이는 '쿠팡' 변수가 관중 수에 대해 통계적으로 유의미한 영향을 미친다는 것을 나타냅니다.

[0.025 0.975]: 이는 '쿠팡' 변수의 계수에 대한 95% 신뢰 구간을 의미합니다. 신뢰 구간은 6266.427에서 1.19e+04 사이로, 이 구간 안에 진짜 계수가 있을 확률이 95%라는 것을 나타냅니다.

결론적으로, 이 회귀분석은 '쿠팡' 활동이 K 리그 1의 관중 수에 긍정적인 영향을 미치는 것으로 나타났으나, 전체 변동성을 크게 설명하지는 못하는 것으로 해석됩니다.
