```python
import pandas as pd
spec_p = pd.read_csv('spec_prep.csv')
spec_p
```

```python
k1 = spec_p[spec_p['리그']=='K LEAGUE 1']
k2 = spec_p[spec_p['리그']=='K LEAGUE 2']
```

```python
import matplotlib.pyplot as plt
# '연도'와 '라운드'를 결합하여 '연도-라운드' 칼럼을 생성
k1['연도-라운드'] = k1['연도'].astype(str) + "-" + k1['라운드'].astype(str)
k2['연도-라운드'] = k2['연도'].astype(str) + "-" + k2['라운드'].astype(str)
# '연도-라운드'를 인덱스로 설정
k1.set_index('연도-라운드', inplace=True)
k2.set_index('연도-라운드', inplace=True)
# matplotlib 폰트 설정
# plt.rcParams['font.family'] = 'NanumGothic'
# # 시계열 데이터 시각화
# plt.figure(figsize=(40, 12))
# plt.plot(k1['관중수'], marker='o', linestyle='-')
# plt.title('연도 및 라운드별 관중 수 추이')
# plt.xlabel('연도-라운드')
# plt.ylabel('관중 수')
# plt.xticks(rotation=45)  # x축 라벨 회전
# plt.grid(True)
# plt.show()
```

```python
import statsmodels.api as sm
import numpy as np
k1['쿠팡'] = k1['연도'].apply(lambda x: 1 if int(x) > 2023 else 0)
k2['쿠팡'] = k2['연도'].apply(lambda x: 1 if int(x) > 2023 else 0)
```

```python
# '연도'와 '라운드'를 정수형으로 변환
k1['연도'] = k1['연도'].astype(int)
k1['라운드'] = k1['라운드'].astype(int)

# '연도'와 '라운드'를 기준으로 데이터프레임을 정렬
k1_ci_sorted = k1.sort_values(by=['연도', '라운드'])

# 다시 평균 관중수를 계산하고, '연도-라운드' 형태로 컬럼을 합친 후, 정렬된 데이터프레임을 생성
k1_group = k1_ci_sorted.groupby(['연도', '라운드'])[['관중수']].mean().reset_index()
k1_group['연도-라운드'] = k1_group['연도'].astype(str) + '-' + k1_group['라운드'].astype(str)
k1_group = k1_group[['연도-라운드','관중수']]
# grouped_sorted.set_index('연도-라운드', inplace=True)
# '연도-라운드' 열에서 '연도'만 추출
k1_group['연도'] = k1_group['연도-라운드'].apply(lambda x: int(x.split('-')[0]))

# '쿠팡' 열 추가 및 조건에 따른 값 부여
k1_group['쿠팡'] = k1_group['연도'].apply(lambda x: 1 if x >= 2023 else 0)
```

```python
# '연도'와 '라운드'를 정수형으로 변환
k2['연도'] = k2['연도'].astype(int)
k2['라운드'] = k2['라운드'].astype(int)

# '연도'와 '라운드'를 기준으로 데이터프레임을 정렬
k2_ci_sorted = k2.sort_values(by=['연도', '라운드'])

# 다시 평균 관중수를 계산하고, '연도-라운드' 형태로 컬럼을 합친 후, 정렬된 데이터프레임을 생성
k2_group = k2_ci_sorted.groupby(['연도', '라운드'])[['관중수']].mean().reset_index()
k2_group['연도-라운드'] = k2_group['연도'].astype(str) + '-' + k2_group['라운드'].astype(str)
k2_group = k2_group[['연도-라운드','관중수']]
# grouped_sorted.set_index('연도-라운드', inplace=True)
# '연도-라운드' 열에서 '연도'만 추출
k2_group['연도'] = k2_group['연도-라운드'].apply(lambda x: int(x.split('-')[0]))

# '쿠팡' 열 추가 및 조건에 따른 값 부여
k2_group['쿠팡'] = k2_group['연도'].apply(lambda x: 1 if x >= 2023 else 0)
```

```python
import numpy as np
import pandas as pd
from causalimpact import CausalImpact

# 데이터 생성
np.random.seed(22)
n_points = 100
x = np.random.normal(0, 0.5, n_points)
y = 2 + 2 * x + np.random.normal(0, 0.5, n_points)
y_intervention = y.copy()
y_intervention[70:] += 5  # 70번째 데이터 포인트 이후에 인과 관계가 있는 개입 발생

data = pd.DataFrame({'y': y_intervention, 'x': x}, columns=['y', 'x'])

# 인과 관계 분석을 위한 기간 설정
pre_period = [0, 69]
post_period = [70, 99]

# CausalImpact 모델 실행
ci = CausalImpact(data, pre_period, post_period)
print(ci.summary())
print(ci.plot())
```
