# 관중수 데이터 전처리

```python
import pandas as pd
spec = pd.read_csv('spec_prep.csv')
spec
```

```python
k1 = spec[spec['리그']=='K LEAGUE 1']
k2 = spec[spec['리그']=='K LEAGUE 2']
```

```python
import numpy as np
k1['쿠팡'] = k1['연도'].apply(lambda x: 1 if int(x) > 2023 else 0)
k2['쿠팡'] = k2['연도'].apply(lambda x: 1 if int(x) > 2023 else 0)
```

```python
k1
```

```python
# '연도'와 '라운드'를 정수형으로 변환
k1['연도'] = k1['연도'].astype(int)
k1['라운드'] = k1['라운드'].astype(int)

# '연도'와 '라운드'를 기준으로 데이터프레임을 정렬
k1_ci_sorted = k1.sort_values(by=['연도', '라운드'])

# 다시 평균 관중수를 계산하고, '연도-라운드' 형태로 컬럼을 합친 후, 정렬된 데이터프레임을 생성
k1_group = k1_ci_sorted.groupby(['연도', '라운드'])[['관중수']].mean().reset_index()
k1_group['연도-라운드'] = k1_group['연도'].astype(str) + '-' + k1_group['라운드'].astype(str)
k1_group = k1_group[['연도-라운드','관중수']]
# grouped_sorted.set_index('연도-라운드', inplace=True)
# '연도-라운드' 열에서 '연도'만 추출
k1_group['연도'] = k1_group['연도-라운드'].apply(lambda x: int(x.split('-')[0]))

# '쿠팡' 열 추가 및 조건에 따른 값 부여
k1_group['쿠팡'] = k1_group['연도'].apply(lambda x: 1 if x >= 2023 else 0)
```

```python
# '연도'와 '라운드'를 정수형으로 변환
k2['연도'] = k2['연도'].astype(int)
k2['라운드'] = k2['라운드'].astype(int)

# '연도'와 '라운드'를 기준으로 데이터프레임을 정렬
k2_ci_sorted = k2.sort_values(by=['연도', '라운드'])

# 다시 평균 관중수를 계산하고, '연도-라운드' 형태로 컬럼을 합친 후, 정렬된 데이터프레임을 생성
k2_group = k2_ci_sorted.groupby(['연도', '라운드'])[['관중수']].mean().reset_index()
k2_group['연도-라운드'] = k2_group['연도'].astype(str) + '-' + k2_group['라운드'].astype(str)
k2_group = k2_group[['연도-라운드','관중수']]
# grouped_sorted.set_index('연도-라운드', inplace=True)
# '연도-라운드' 열에서 '연도'만 추출
k2_group['연도'] = k2_group['연도-라운드'].apply(lambda x: int(x.split('-')[0]))

# '쿠팡' 열 추가 및 조건에 따른 값 부여
k2_group['쿠팡'] = k2_group['연도'].apply(lambda x: 1 if x >= 2023 else 0)
```

```python
k1_group
```

```python
import pandas as pd
k1_group = pd.read_csv('k1_group.csv')
```

```python
# # 새로운 데이터
new_data = {'관중수': [16572.8], '연도': [2024], '쿠팡': [1]}

# # 데이터프레임 생성
new_row = pd.DataFrame(new_data)

# # 기존 데이터프레임에 새로운 행 추가
k1_group = pd.concat([k1_group, new_row], ignore_index=True)
```

```python
result = adfuller(k1_group['관중수'])

print(f'ADF statistics : {result[0]}')
print(f'p-value {result[1]}')
print('critical values: ')
for key, value in result[4].items():
    print(f"\t{key} {value}")
```

```python
if result[1] < 0.05:
    print("reject the null hypothesis - the time series doesn't have unit root, i.e it is stationary")
else:
    print("fail to reject the null hypothesis - i.e it is non-stationary")
```

```python
# Causal Impact

from causalimpact import CausalImpact
# '연도', '월', '일' 컬럼을 사용하여 날짜 인덱스 생성

# 분석에 사용할 변수 선택 (여기서는 '관중수')
analysis_data = k1_group[['관중수_diff']]

# 분석 기간 설정: 'pre_period'는 이벤트 전, 'post_period'는 이벤트 후 시간을 정의
pre_period = [0, 174]
post_period = [175, 253]

# CausalImpact 분석 수행
ci = CausalImpact(analysis_data, pre_period, post_period)
print(ci.summary())
```

```python
import statsmodels.api as sm
# '관중수'에 대한 회귀분석 준비
X = sm.add_constant(k1_group['쿠팡'])  # 상수항 추가 및 '쿠팡'이 개입 변수 역할을 합니다.
y = k1_group['관중수_diff']

# OLS 회귀모델 구축 및 결과 출력
model = sm.OLS(y, X).fit()
print(model.summary())
```
