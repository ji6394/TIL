```python
import pandas as pd
from eunjeon import Mecab

# MeCab 객체 초기화
mecab = Mecab()

# 형태소 분석 함수 정의
def analyze_text(text):
    parsed = mecab.nouns(text)
    return parsed

# 'subject'와 'contents' 열에 대한 형태소 분석 수행
RS_posts['subject_parsed'] = RS_posts['subject'].apply(analyze_text)
RS_posts['contents_parsed'] = RS_posts['contents'].apply(analyze_text)

print(RS_posts)
```
