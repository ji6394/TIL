```python
k1 = spec[spec['리그']=='K LEAGUE 1']
k2 = spec[spec['리그']=='K LEAGUE 2']
```

```python
# 각 연도별로 '라운드' 열의 유니크한 값들을 확인
for year in k2['연도'].unique():
    unique_rounds = k2[k2['연도'] == year]['라운드'].unique()
    unique_rounds.sort()  # 결과를 보기 쉽게 정렬
    print(f"{year}년도의 라운드: {unique_rounds}")
```
