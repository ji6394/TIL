```python
import random
import numpy as np
import os
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestClassifier
```

```python
def seed_everything(seed):
    random.seed(seed)
    os.environ['PYTHONHASHSEED'] = str(seed)
    np.random.seed(seed)
seed_everything(7) # Seed 고정
```

```python
import pandas as pd
train = pd.read_csv('/content/drive/MyDrive/LG/train.csv')
test = pd.read_csv('/content/drive/MyDrive/LG/test.csv')
```

```python
train=train.fillna(0)
test=test.fillna(0)
```

```python
train_x = train.drop(columns=['PRODUCT_ID','Y_Class','Y_Quality','TIMESTAMP'])
test_x = test.drop(columns=['PRODUCT_ID','TIMESTAMP'])
train_x_num = train_x.filter(regex='X')
test_x_num = test_x.filter(regex='X')
#min-max-scaling이 더 나을듯
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(train_x_num)
train_x_num_scaled = scaler.transform(train_x_num)
test_x_num_scaled = scaler.transform(test_x_num)
train_x_num_scaled = pd.DataFrame(train_x_num_scaled)
test_x_num_scaled = pd.DataFrame(test_x_num_scaled)
```

```python
train_x_num_scaled
```

```python
test_x_num_scaled
```

```python
#주성분분석(연속형변수)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=598)
pca_array = pca.fit_transform(train_x_num_scaled)
pca_df = pd.DataFrame(pca_array)
pca_df.head()
```

```python
result = pd.DataFrame({'설명가능한 분산 비율(고윳값)':pca.explained_variance_, '기여율':pca.explained_variance_ratio_})
result['누적기여율']=result['기여율'].cumsum()
result
```

```python
#3개만하자
from sklearn.decomposition import PCA
pca = PCA(n_components=3)
train_num_pca = pca.fit_transform(train_x_num_scaled)
train_pca=pd.DataFrame(data=train_num_pca)
```

```python
train_pca
```

```python
test_num_pca = pca.transform(test_x_num_scaled)
test_pca=pd.DataFrame(data=test_num_pca)
```

```python
test_pca
```

```python
cat_encoding_train = pd.get_dummies(train_x[['LINE','PRODUCT_CODE']])
cat_encoding_test = pd.get_dummies(test_x[['LINE','PRODUCT_CODE']])
```

```python
train_x_prep = train_pca.join(cat_encoding_train)
test_x_prep = test_pca.join(cat_encoding_test)
train_y = train[['Y_Quality','Y_Class']]
```

```python
train_x_prep
```

```python
test_x_prep
```

```python
from sklearn.model_selection import train_test_split
x_train, x_val, y_train, y_val = train_test_split(train_x_prep, train_y, test_size = 0.1, random_state = 7)
```

```python
y_train_qual = y_train['Y_Quality']
y_val_qual = y_val['Y_Quality']
y_train_cl = y_train['Y_Class']
y_val_cl = y_val['Y_Class']
```

```python
#RandomforestRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
rfr = RandomForestRegressor()
rfr.fit(x_train, y_train_qual)
rfr_pred = rfr.predict(x_val)
mae = mean_absolute_error(y_val_qual, rfr_pred)
mse = mean_squared_error(y_val_qual, rfr_pred)
print('MAE: %d' %mae)
print('MSE: %d' %mse)
print(r2_score(y_val_qual, rfr_pred))
```

```python
# train_class = pd.DataFrame(train['Y_Class'])
# train_qual = pd.DataFrame(train['Y_Quality'])
cq_rf = RandomForestClassifier(random_state=7).fit(pd.DataFrame(y_train_qual), pd.DataFrame(y_train_cl))
```

```python
cq_preds = cq_rf.predict(pd.DataFrame(rfr_pred))
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_val_cl, cq_preds))
```

```python
#XGBoost
from xgboost import XGBRegressor
xgr = XGBRegressor()
xgr.fit(x_train, y_train_qual)
xgr_pred = xgr.predict(x_val)
mae = mean_absolute_error(y_val_qual, xgr_pred)
mse = mean_squared_error(y_val_qual, xgr_pred)
print('MAE: %d' %mae)
print('MSE: %d' %mse)
print(r2_score(y_val_qual, xgr_pred))
```
